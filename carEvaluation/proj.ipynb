```python

file = open('car.csv', 'r')

# Class Values: 

# unacc, acc, good, vgood 

# Attributes: 

# buying: vhigh, high, med, low. 
# maint: vhigh, high, med, low. 
# doors: 2, 3, 4, 5more. 
# persons: 2, 4, more. 
# lug_boot: small, med, big. 
# safety: low, med, high. 

buying = {
            'vhigh':'20',
            'high':'15',
            'med':'10',
            'low':'5'
            }
maint = {
            'vhigh':'20',
            'high':'15',
            'med':'10',
            'low':'5',
        }
doors = {
            '2':'2',
            '3':'3',
            '4':'4',
            '5more':'5'
          }

persons = {
            '2':'2',
            '4':'4',
            'more':'8'
            }
lug_boot ={ 
            'small':'2',
            'med':'4',
            'big':'8'
             }

safety = {
            'low':'2',
            'med':'4',
            'high':'8'
           }
# unacc, acc, good, vgood 
classe = {
            'unacc':'2',
            'acc':'4',
            'good':'8',
            'vgood':'16'
        }

df = pd.read_csv('car.csv',header=0)


def handle_non_numerical_data(df):
    columns = df.columns.values
    
    for column in columns:
        text_digic_vals = {}
        
        def convert_to_int(val):
            return text_digic_vals[val]
        
        if df[column].dtype != np.int64 and df[column].dtype != np.float64:
            column_contents = df[column].values.tolist()
            unique_elements = set(column_contents)
            
            x=0
            for unique in unique_elements:
                if unique not in text_digic_vals:
                    text_digic_vals[unique]=x
                    x+=1
            
            df[column] = list(map(convert_to_int,df[column]))
    
    return df

df = handle_non_numerical_data(df)

X = np.array(df.drop('classe',1).astype(int))
Y = np.array(df['classe'])

pca = PCA(n_components=6)
X_spca = pca.fit_transform(X)

fig, ax = plt.subplots(nrows=1728, ncols=6, figsize=(5, 5))
ax=ax.flatten()
ax[0].scatter(X_spca[Y == 0, 0], X_spca[Y == 0, 1],
              color='red', marker='^', alpha=0.5)
ax[0].scatter(X_spca[Y == 1, 0], X_spca[Y == 1, 1],
              color='blue', marker='o', alpha=0.5)
inx=sum(Y==0)
print(inx)
ax[1].scatter(X_spca[Y == 0, 0], np.zeros((inx, 1)) + 0.02,
              color='red', marker='^', alpha=0.5)
inx=sum(Y==0)
print(inx)
ax[1].scatter(X_spca[Y == 1, 0], np.zeros((inx, 1)) - 0.02,
              color='blue', marker='o', alpha=0.5)

ax[0].set_xlabel('PC1')
ax[0].set_ylabel('PC2')
ax[1].set_ylim([-1, 1])
ax[1].set_yticks([])
ax[1].set_xlabel('PC1')

plt.tight_layout()
# plt.savefig('./figures/half_moon_2.png', dpi=300)
plt.show()

```
